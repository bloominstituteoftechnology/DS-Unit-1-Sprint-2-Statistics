# Apply the t-test to real data

Your assignment is to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

Your goals:

1. Load and clean the data (or determine the best method to drop observations when running tests)
2. Using hypothesis testing, find an issue that democrats support more than republicans with p < 0.01
3. Using hypothesis testing, find an issue that republicans support more than democrats with p < 0.01
4. Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1 (i.e. there may not be much of a difference)

Note that this data will involve *2 sample* t-tests, because you're comparing averages across two groups (republicans and democrats) rather than a single group against a null hypothesis.

```python
# Walt Perris

# imports/setups
import numpy as np
import pandas as pd
from scipy import stats

# read the file
column_headers = ['party','handicapped-infants','water-project','budget',
                  'physician-fee-freeze','el-salvador-aid','religious-groups',
                  'anti-satellite-ban','aid-to-contras','mx-missile',
                  'immigration','synfuels','education','right-to-sue','crime',
                  'duty-free','south-africa']
 
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data", names=column_headers)

print(df.shape)
df.head()
```

```python
df.describe() 
```

```python
df.isnull().sum()
```

```python
# Clean
df = df.replace({"y": 1, "n": 0, "?": np.NaN})

print(df.shape)
df.head()
```

```python
# Republican

rep = df[df['party']=='republican']
rep.head()
```

```python
# Democrat

dem = df[df['party']=='democrat']
dem.head()
```

```python
stats.ttest_ind(rep['education'], dem['education'], nan_policy='omit')

# Based on a t-statistic of ~20 and a p-value of ~0. I reject the null hypothesis 
# that republicans and democrats support the 'education' bill at the same level/rate. 
# And I suggest the alternative that democrat and republican levels of support for 
# this bill are different.
```

```python
stats.ttest_ind(rep['religious-groups'], dem['religious-groups'], nan_policy='omit')

# Based on a t-statistic of 9.73 and a p-value of 1.88 I fail to reject the null hypothesis
# that republicans and democrats support the 'religious-groups' bill at the same level/rate
# which suggests the levels of support from the democrats and republicans are within
# similar levels/rates. 
```

```python
stats.ttest_ind(rep['synfuels'], dem['synfuels'], nan_policy='omit')

# Based on a t-statistic of 8.29 and a p-value of 2.39 I fail to reject the null hypothesis
# that republicans and democrats support the 'synfuels' bill at the same level/rate
# which suggests the levels of support from the democrats and republicans are within
# similar levels/rates.
```
