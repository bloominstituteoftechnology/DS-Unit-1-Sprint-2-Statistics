It's good you asked, because I spent quite some time on that very question. There's a lot to say about it. Like Frequentist as a rule observe things as they happen, and reacting to it. A Bayesian will form an opinion and try to see if their opinion is correct as the input comes in. The Bayesian approach in my understanding is more biased than the a frequentist who will generally form no bias or anything to disprove and will simply allow the observation to form the definition. Bayesian also seems to be more theoretical approach to probability theory, where frequentism appears more applied. Why? Because the questions one could pose about a situation before executing a test is the Bayesian method. This is sympomatic of philosophy which does not require empirical observation of their theories. This suspicion is further confirmed by the fact that Bayesian theory was popular before the 20th century. And frequentist is more popular after 20th century. Bayesian is similar to an AGI (artificial general intelligence) consciousness contemplating on what action would be ideal to take on the physical world, without ever actually had taken any physical action. Frequentist on the other side would be an AGI taking action on its physical environment and forming ideas about what is the best next action due to its experience. On the other hand as it's name suggest, frequentists define probability based on the frequency which something happened, or 'long-term frequency'. Also they have a fundamental belief in predicting using patterns. But Bayesian probability is based in degrees of logical support (then belief).
As a long term practitioner of probabilities for my own business needs I would then agree that frequentist probability is a simpler, lower level, and Bayesian is a higher level which could easily outmaneuver frequentist probability.
