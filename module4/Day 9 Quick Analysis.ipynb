```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
%matplotlib inline
g = sns.PairGrid(df)
g.map(plt.scatter)
```

```
g = sns.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);
```

```
# choose one column from dataset to set the hue
g = sns.PairGrid(df, hue="airline")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```

```
#  compare two numerical columns, adjusting hue by airline column
g = sns.PairGrid(df, vars=["incidents_00_14", "fatalities_00_14"], hue="airline")
g.map(plt.scatter);
```

```
g = sns.PairGrid(df)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False);
```
