```
# TODO - code!
# A function to apply bayes theorem to drug test.

def bayes(prior, false_positive_rate):
  '''
    Bayes Theorem Function

    # Generic Formula

    P(User | +)= P(+|User)*P(User) / P(+) or P(A|B)=P(B|A)P(A)/P(B)

    Above formula tranlation in relation to testing positive in alchol test:

      P(A|B)- The probability of A given situation B (probability
        of users(A) that will test positive on alcohol test(B = situation)) = 

      (P(B|A)-The percentage of users who test positive on alcohol test *
        P(A) -the probability of somebody testing positive on alcohol test) /  
      P(B) - The probability of the situation/event happening

      P(B) translates to:
      (P(B|A)-The percentage of users who test positive on alcohol test *
        P(A)-the probability of somebody testing positive on alcohol test) + 
      (the percentage of people that test false positive on alcohol test *
        the probability of somebody testing negative on alcohol test )

    detecting how many times the machine accurately 
      positively identifies a user

    # Overall Formula for this function assuming it is used to calculate
    the accuracy of an alcohol detection machine. 

    (true positive user rate * prior)/ 
      ((true positive  user rate * prior))+ 
      (false positive non user * non-user) 

    prior/user rate = 1/200 = 0.005 - what percentage 
      of population are users

    true positive user rate = 99% = 0.99 - Machine accurately  
      positively detects user 99% of time

    true positive user rate = 1-false positive = 1 - 0.01 = 0.99 - percentage
      of people who were accurately detected to be a user. 

    false positive non user rate= 0.01 = 1% - machine falsely 
      detect a non user as a user 1% of time

    non-user = 1-user = 1-0.005 = 0.995 - percentage of 
      population that is not a user. 

  '''

  true_positive_rate = 1 - false_positive_rate
  non_user = 1 - prior

  probability = (true_positive_rate * prior) / ((true_positive_rate * prior) + 
                                  (false_positive_rate * non_user))
  
  bayes_theory = probability * 100

  return bayes_theory

# Testing the bayes function
bayes(0.005, 0.01)

# My Bayes theory function works. 
```

```
from scipy import stats
import numpy as np
import pandas as pd
pd.options.display.max_rows = 150
pd.options.display.max_columns = 150

loans = pd.read_csv('LoanStats_2018Q4.csv', header = 1, skipfooter=2, engine='python')

def confidence_interval(data, confidence=0.95):
  ''' Calculate a confidence interval around a sample mean for given data.
  Using t-distribution and two-tailed test, default 95% confidence. 
  
  Arguments:
    data - iterable (list or numpy array) of sample observations
    confidence - level of confidence for the interval
  
  Returns:
    tuple of (mean, lower bound, upper bound)'''

  data=np.array(data)
  mean=np.mean(data)
  n=len(data)
  stderr=stats.sem(data)
  margin=stderr * stats.t.ppf((1+confidence) / 2.0, n-1)
  return (mean, mean-margin, mean+margin)
```

```
loans = loans.reset_index()
```

```
# Confidence interval of the frequentist using annual income data
frequent = confidence_interval(loans['annual_inc'])
frequent
```

```
#confidence interval of the bayesian using annual income data
bay_mean, bay_var, bay_std = stats.bayes_mvs(loans['annual_inc'], alpha=0.95)
bay_mean
```

```

import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots()
fig.patch.set(facecolor='white')

# Plotting the mean and confidence interval of the frequentist
# statistics data for annual income  


sns.kdeplot(loans['annual_inc'])
CI = confidence_interval(loans['annual_inc'])
plt.axvline(x=CI[1], color='k') #minimum bound # one here means index 1 of the confidence 
# interval result.  So like saying confidence_interval[1] 
# which means give me the result located at index 1 of the list of results
# from the confidence interval. The result located in the index 1 represents
# the lower bound of the confidence interval. index 2 below is upper bound of
# confidence interval, index 0 below is the mean of the data set. 
plt.axvline(x=CI[2], color='blue') # max bound
plt.axvline(x=CI[0], color='red') # mean
plt.title('Average Annual income of loan customers using Frequentist Method');
plt.show()

# Plotting the mean and confidence interval of the bayesian
# statistics data for annual income  

fig, ax = plt.subplots()
fig.patch.set(facecolor='white')
sns.kdeplot(loans['annual_inc'])
bay_mean, bay_var, bay_std = stats.bayes_mvs(loans['annual_inc'], alpha=0.95)
plt.axvline(x=bay_mean[1][0], color='k') # minimum bound
plt.axvline(x=bay_mean[1][1], color='blue') # max bound
plt.axvline(x=bay_mean[0], color='red') # mean
plt.title('Average Annual income of loan customers using Bayesian Method');

plt.show()

```

```
# Describe Bayesian vs frequentist in your own words

'''Bayesian statistics uses information that was previously known
in its calculations. Where this prior information came from has no 
universal basis, sometimes it is arbitrary. The bayesian might update
their believes based on the results of the experiment. They repeat
their experiment multiple times to build a higher confidence level
and the "prior" going into every new experiment is the result derived 
from the previous experiment. They use subsequent experiment to 
validate and gain higher confidence on their prior beleive. 

Frequentist approach bases their conclusions purely on the results
of the experiment. They repeat the experiment multiple times to prove
whether the results of the experiments are significant or not. 
They do not use the results of the previous experiment as a basis
for the next experiment. Every experiment is conducted independent
of the results of the last experiment. They use the result of subsequent
experiments to prove the first experiment is accurate and significant
but not for the purpose of validating their prior beleives.'''
```
