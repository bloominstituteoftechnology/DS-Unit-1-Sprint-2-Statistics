```

def drunken_baye(true_positive, prior, false_positive_rate):
  prior_compliment = float(1 - prior)
  numerator = float(true_positive * prior)
  denominator = float((true_positive)*(prior) + (false_positive_rate)*(prior_compliment))
  posterior = numerator/denominator
  return posterior
```

```
#What is the probability that a person is drunk after one positive breathalyzer test?

Posterior_1 = drunken_baye(1,.1/1000,.08)

Posterior_1
```

```
#What is the probability that a person is drunk after two positive breathalyzer tests?
Posterior_2 = drunken_baye(1,Posterior_1,.08)

Posterior_2
```

```
Posterior_3 = drunken_baye(1,Posterior_2,.08)

Posterior_3
```

```
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
```

```
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
```

```
df.head()
```

```
df_sub = df['respondents']


df_sub = np.array(df_sub)

df_sub
```

```
#frequentist approach

def confidence_interval(data, confidence=.95):
  n = len(data)
  mean = sum(data)/n
  data = np.array(data)
  stderr = stats.sem(data)
  interval = stderr * stats.t.ppf((1 + confidence) / 2.0, n-1)
  return (mean, (mean-interval, mean+interval))

FR = confidence_interval(df_sub)

print(FR)
```

```
#bayesian approach

CI, _, _ = stats.bayes_mvs(df_sub, alpha=.95)

CI
```

```
new_df = pd.DataFrame({'Mean': [FR[0],CI[0]],
                       'Lower_Bound':[CI[1][0], FR[1][0]],
                       'Upper_Bound':[CI[1][1],FR[1][1]],
                       })
```

```
new_df['Margin_of_Error'] = new_df['Upper_Bound'] - new_df['Mean']

new_df
```

```
plt.bar(x=range(0,2),height=new_df['Mean'],yerr=new_df['Margin_of_Error'])
plt.xticks([0,1],labels=['Frequentist', 'Bayesian'])
plt.title('Frequentist vrs Bayesian Margin of Error')
plt.ylim(375,410)
plt.show()
```
