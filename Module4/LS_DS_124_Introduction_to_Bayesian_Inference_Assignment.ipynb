### 1) Write a function 

`def prob_drunk_given_positive(prob_drunk_prior, false_positive_rate, true_positive_rate):` 

You should only truly need these three values in order to apply Bayes Theorem. In this example, imagine that individuals are taking a breathalyzer test with an 8% false positive rate, a 100% true positive rate, and that our prior belief about drunk driving in the population is 1/1000. 
 - What is the probability that a person is drunk after one positive breathalyzer test?
 - What is the probability that a person is drunk after two positive breathalyzer tests?
 - How many positive breathalyzer tests are needed in order to have a probability that's greater than 95% that a person is drunk beyond the legal limit?


```
#let's write the function
def prob_drunk_given_positive(prob_drunk_prior, false_positive_rate, true_positive_rate):
  # bayes formula
  numerator = true_positive_rate * prob_drunk_prior
  marginal =  (true_positive_rate * prob_drunk_prior) + (false_positive_rate * (1 - prob_drunk_prior))
  posterior1 = numerator / marginal 
  return posterior1
```

```
#false_positive_rate=.08(8%)
# prob_drunk_prior = 1/1000
# true positive rate = 1 (100%)
posterior1 = prob_drunk_given_positive(0.001, 0.08, 1)
print(posterior1)
print('this is one positive breathalyzer test')
```

```
posterior2 = prob_drunk_given_positive(posterior1, 0.08, 1)
print(posterior2)
print('this is two positive breathalyzer test')
```

```
posterior3 = prob_drunk_given_positive(posterior2, 0.08, 1)
print('this is three positive breathalyzer test')
print(posterior3)
```

```
posterior4 = prob_drunk_given_positive(posterior3, 0.08, 1)
print('this is four positive breathalyzer test')
print(posterior4)
```

```
#So four breathalyzer test is needed in order to have a probablity greater than 95%
```
