```python
import sys

# If you're on Colab:
if 'google.colab' in sys.modules:
    DATA_PATH = 'https://raw.githubusercontent.com/LambdaSchool/DS-Unit-2-Applied-Modeling/master/data/'

# If you're working locally:
else:
    DATA_PATH = '../data/'
    
# Ignore this Numpy warning when using Plotly Express:
# FutureWarning: Method .ptp is deprecated and will be removed in a future version. Use numpy.ptp instead.
import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning, module='numpy')
```

```python
# Read New York City apartment rental listing data
import pandas as pd
df = pd.read_csv(DATA_PATH+'apartments/renthop-nyc.csv')
assert df.shape == (49352, 34)
```

```python
# Remove outliers: 
# the most extreme 1% prices,
# the most extreme .1% latitudes, &
# the most extreme .1% longitudes
df = df[(df['price'] >= 1375) & (df['price'] <= 15500) & 
        (df['latitude'] >=40.57) & (df['latitude'] < 40.99) &
        (df['longitude'] >= -74.1) & (df['longitude'] <= -73.38)]
```

```python
# Exploring data

df.head()
```

```python
# Exploring numeric data

df.describe()
```

```python
import plotly.express as px
```

```python
# Visually exploring data to find trendline with highest R2 value

px.scatter(df, x='bathrooms', y='price', trendline='ols')
```

```python
# Importing Linear Regression model from SKLearn

from sklearn.linear_model import LinearRegression
```

```python
# Instantiating model

model = LinearRegression()
```

```python
# Defining x and y variables for training set

features = ['bathrooms']
target = ['price']

X_train = df[features]
y_train = df[target]
```

```python
# Fitting model to training set

model.fit(X_train, y_train)
```

```python
# Applying model to new data

bathrooms = 3
x_test = [[ bathrooms ]]

y_pred = model.predict(x_test)
y_pred
```

```python
# Defining function to predict price based on number of bathrooms in apartment.

def predict(bathrooms):
  y_pred = model.predict([[bathrooms]])
  estimate = y_pred[0]
  print(int(estimate), "is our estimated monthly rent for a NYC apartment with",
        int(bathrooms), "bathrooms.")

predict(3)
```

```python
print(model.coef_)
```

The model coefficient for this linear regression model implies that, for every additional bathroom in a New York City apartment, the monthly rent for said apartment increases by $2,573.37. This is the slope of the line represented by our linear regression model.
