## Normal Distribution Revisited

What is the Normal distribution: A probability distribution of a continuous real valued random-variable. The Normal distribution properties make it useful for the *Central Limit Theorm*, because if we assume a variable follows the normal distribution, we can make certain conclusions based on probabilities.

```python
import numpy as np

mu1 = 00 # mean
sigma1 = 1 # standard deviation

sample1 = np.random.normal(mu1, sigma1, 1000)
sample1
```

```python
# Verify the variance of our sample
abs(sigma1 - np.std(sample1, ddof=1)) < 0.01
```

### Helpful video on why we use the t-distribution

<https://www.youtube.com/watch?v=Uv6nGIgZMVw>

However, in order to understand it you'll need to understand what a z-score is:

A z-score calculates the number of standard deviations an observations lies from the population mean. The problem is that in real-world situations, we don't know what the sample mean is, so we have to turn to using the sample mean to estimate the population mean. Because the sample mean is generated from a sample and used to estimate the population mean with some level of uncertainty, it also has its own distribution a nd spread. This means that for low sample sizes both our estimates of the sample mean and sample population are not very precise, they're kind of spread out. It's this spread that makes the t-distribution wider than the normal distribution for low sample sizes. However, with the larger the sample size, the closer the t-distribution approximates the normal distribution.


## Student's T Test

>Assuming data come from a Normal distribution, the t test provides a way to test whether the sample mean (that is the mean calculated from the data) is a good estimate of the population mean. 

The derivation of the t-distribution was first published in 1908 by William Gosset while working for the Guinness Brewery in Dublin. Due to proprietary issues, he had to publish under a pseudonym, and so he used the name Student.

The t-distribution is essentially a distribution of means of normaly distributed data. When we use a t-statistic, we are  checking that a mean fails within a certain $\alpha$ probability of the mean of means.

```python
t_df10 = np.random.standard_t(df=10, size=10)
t_df100 = np.random.standard_t(df=100, size=100)
t_df1000 = np.random.standard_t(df=1000, size=1000)
```

```python
sns.kdeplot(t_df10, color='r');
sns.kdeplot(t_df100, color='y');
sns.kdeplot(t_df1000, color='b');
```

```python
i = 10
for sample in [t_df10, t_df100, t_df1000]:
    print(f"t - distribution with {i} degrees of freedom")
    print("---" * 10)
    print(f"Mean: {sample.mean()}")
    print(f"Standard Deviation: {sample.std()}")
    print(f"Variance: {sample.var()}")
    i = i*10
```

Why is it different from normal? To better reflect the tendencies of small data and situations with unknown population standard deviation. In other words, the normal distribution is still the nice pure ideal (thanks to the central limit theorem), but the t-distribution is much more useful in many real-world situations.

```python
import numpy as np
mu = 0    # numerical average (mean)
sigma = 1 # standard deviation

# import a random sample of 1000
sample = np.random.normal(mu, sigma, 1000)

# view sample's first 5
sample[:5]
```

```python
# Generate a Hypothesis

# Null Hypothesis(boring ): Sample A & Sample B are the same, split 50/50
# Alternative hypothesis (1-tailed): Sample A has a lower/higher conversion rate than sample B
# Alternative hypothesis (2-tailed): Sample A has a different conversion rate than sample B
```

```python
##### Example sample data - (conversion rates of a website over 10 days)
a=[1136,1178,1212,1193,1226,1154,1230,1222,1161,1148]
b=[1184,1203,1219,1238,1243,1204,1269,1256,1156,1248]

print(np.mean(a))
print(np.mean(b))
```

```python
# import
from scipy.stats import ttest_ind

# compare
# This gives us a statistic
print(ttest_ind(a,b))
print(ttest_ind(a,b).statistic)
```

```python
# THE P-VALUE
# The degree of confidence that the two sample groups are different from one another.

# 95% confidence interval (alpha = .05)
# 1-TAIL TEST
print('Confidence Interval = 95%')
print(ttest_ind(a,b).pvalue < .05) #### RESULT TRUE BECAUSE IT'S pvalue=0.03504505626727664
# Is the P value less than a .05 
# or in other words, is my confidence interval greater than 95%?
# 2-Tail test
print(ttest_ind(a,b).pvalue/2 < .05) #### TRUE
print('I reject the null hypothesis')

# 99% confidence interval (alpha = .01)
print('\nConfidence Interval = 99%')
print(ttest_ind(a,b).pvalue < .01) #### FALSE
print(ttest_ind(a,b).pvalue/2 < .01) #### FALSE
# The P value is less than .01, causing the result to be false.
# The confidence interval is not greater than 99%
print('I cannot reject the null hypothesis')

# 99.9% confidence interval (alpha = .001)
print('\nConfidence Interval = 99.9%')
print(ttest_ind(a,b).pvalue < .01) #### FALSE
print(ttest_ind(a,b).pvalue/2 < .01) #### FALSE
# The P value is less than .001, causing the result to be false.
# The confidence interval is not greater than 99%
print('I cannot reject the null hypothesis')
```

```python
# THE T-STATISTIC

# The numerator is the signal. You simply take the sample mean and subtract the 
# null hypothesis value. As the difference between the sample mean and the null 
# hypothesis mean increases in either a positive or negative direction, the 
# strength of the signal increases.

# The denominator is the noise. The equation in the denominator is a measure of 
# variability known as the Standard of Error of the mean. This statistic indicates 
# how accurately your sample estimates the meaning of the population. A larger number 
# indicates that your sample estimate is less precise because it has more random error.
ttest_ind(a,b).statistic

# Both the signal-to-noise values are in the units of your data. If your signal 
# is six and the noise is to, your t-value is three. This t-value indicates that 
# the difference is three times the size of the standard air. However, if there 
# is a difference of the same size but your data have more variability (six), 
# your t-value is only one. The signal is at the same scale as the noise.
```

## Setting it up

We need three things to start
1. Null hypothesis (the results will be equal)
2. Alternative hypothesis (they will not be the same)
3. Confidence level: The probability of seeing a true result in spite of random variability. (How often do I want to make sure that I'm right.) 
- Typically: 95%, 99%, 99.9%


```python
import scipy.stats # For independant samples
```

```python
# Pancake sample1 (1000 pancakes)
mu1 = 180 # Sample mean / SECONDS OF COOKING
sigma1 = 5 # Standard deviation

# generate sample
sample1 = np.random.normal(mu1, sigma1, 1000)
```

```python
# Pancake sample1 (1000 pancakes)
mu2 = 178.5 # Sample mean / SECONDS OF COOKING
sigma2 = 4.25 # Standard deviation

# generate sample
sample2 = np.random.normal(mu2, sigma2, 800)
```

```python
# compare
results = ttest_ind(sample1, sample2)
print(results)

# Two ways to access the p-value
print(results[1])
print(results.pvalue)
```

```python
# Retreive p-value from scientific notation
print('{:f}'.format(results.pvalue))
print('{:.94f}'.format(results.pvalue))
print('Due to the T-STATISTIC being high(high signal, not much noise) and the p-value being low')
print('(way less than .05 - unlikely to happen by chance), I reject the null hypothesis that the')
print('cooking time is the same and suggest the alternative hypothesis, that they are different.')
print('(Because the p-value was less than .05, we reject the null hypothesis)')
```
