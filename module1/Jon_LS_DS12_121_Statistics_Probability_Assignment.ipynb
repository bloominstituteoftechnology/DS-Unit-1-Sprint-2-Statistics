# Apply the t-test to real data

Your assignment is to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

Your goals:

1. Load and clean the data (or determine the best method to drop observations when running tests)
2. Using hypothesis testing, find an issue that democrats support more than republicans with p < 0.01
3. Using hypothesis testing, find an issue that republicans support more than democrats with p < 0.01
4. Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1 (i.e. there may not be much of a difference)

Note that this data will involve *2 sample* t-tests, because you're comparing averages across two groups (republicans and democrats) rather than a single group against a null hypothesis.

```python
### YOUR CODE STARTS HERE
column_headers = ['party', 'handicapped-infants', 'water-project', 
'budget', 'physician-fee-freeze', 'el-salvador-aid', 'religious-groups', 
'anti-satellite-ban', 'aid-to-contras', 'mx-missile', 'immigration', 
'synfuels', 'education', 'right-to-sue', 'crime', 'duty-free', 
'south-africa']

import pandas as pd
import numpy as np

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data', names= column_headers)
print(df.shape)
df.head()
```

```python
df = df.replace({ '?' : np.NaN, 'n': 0, 'y' : 1})
df.head()
```

```python
df.isnull().sum()
```

```python
rep = df[df.party == 'republican']
rep.head()
```

```python
rep.describe()
```

```python
dem = df[df['party'] == 'democrat' ]
dem.describe()
```

```python
from scipy.stats import ttest_ind, ttest_ind_from_stats, ttest_rel
```

```python
# 2. Find dems more than rep issue
# Ho : The mean of dems.handicapped-infants <= the mean of rep.handicapped-infants
# Ha : The mean of dems.handicapped-infants > the mean of rep.handicapped-infants
ttest_ind(rep['handicapped-infants'], dem['handicapped-infants'], nan_policy='omit')
```

```python
# Since T = -9.205 and p-value/2 = (1.6e-18)/2 = 8e-19, we are 95% confident we
# can reject the null hypothesis that  dems.handicapped-infants are less than or
# equal to rep.handicapped-infants
```

## Since T = -9.205 and p-value/2 = (1.6e-18)/2 = 8e-19, we are 95% confident we can reject the null hypothesis that  dems.handicapped-infants are less than or equal to rep.handicapped-infants. Democrats are more likely to vote yes for the handicapped-infants issue than republicans are.

```python
# 4. Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1 
# Ho : the mean of rep['immigration'] =  the mean of dem['immigration']
# Ha : the mean of rep['immigration'] !=  the mean of dem['immigration']

ttest_ind(rep['immigration'], dem['immigration'], nan_policy='omit')
```
